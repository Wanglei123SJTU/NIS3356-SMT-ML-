```python
import re
import jieba
import cntext as ct

stopwords = ct.load_pkl_dict('STOPWORDS.pkl')['STOPWORDS']['chinese']

def clean_text(text):
    words = jieba.lcut(text)
    words = [w for w in words if w not in stopwords]
    return ' '.join(words)

```

```python
from gensim.models import KeyedVectors

chinese_news_models = KeyedVectors.load_word2vec_format('sgns.wiki.word.bz2', unicode_errors='ignore')
chinese_news_models
```

```python
print(len(data))
```

```python
data_weibo_div=weibo_list[0:857]
print(data_weibo_div)
data_final=[]
for i in data_weibo_div:
    data_final.append(clean_text(i))
```

```python
print(data_final)
```

```python
data_final_div=[]
for i in data_final:
    if(len(i)>=10): data_final_div.append(i)
```

```python
print(len(data_final_div))
```

```python

from umap import UMAP
from hdbscan import HDBSCAN
from sklearn.cluster import KMeans
umap_model = UMAP(n_neighbors=30, n_components=5, min_dist=0, metric='cosine')
# hdbscan_model = HDBSCAN(min_cluster_size=10, min_samples=1,metric='euclidean', cluster_selection_method='eom')
hdbscan_model = HDBSCAN(min_cluster_size=10, metric='euclidean', cluster_selection_method='eom')
```

```python
from bertopic import BERTopic


topic_model = BERTopic(language="chinese (simplified)", 
                       embedding_model=chinese_news_models,
                       umap_model=umap_model,
                       hdbscan_model=hdbscan_model,
                       calculate_probabilities=False, 
                       verbose=True,
                       nr_topics='auto'
                       )


# docs = data['head_div'].tolist()
docs = data_final_div
#2000条进行fit_transform需要1min
topics, probs = topic_model.fit_transform(docs)
```

```python
topic_model.get_topic_info()
```

```python
# 可视化选定的主题和文档
topic_model.visualize_documents(docs)
```

```python
new_topics = topic_model.reduce_outliers(docs, topics)
topic_model.update_topics(docs, topics=new_topics)
```
